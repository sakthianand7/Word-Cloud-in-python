```

import wordcloud
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import display
import io
import sys
```

```
filename=open('/content/output1.txt','r')
file_contents=filename.read()
```

```
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
dont_care_terms = ["the", "a", "to", "if",'in','for','is','through', "it", "of", "and", "or", "an", "as", "i", "me", "my", \
"we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
"their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
"have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
"all", "any", "both", "each", "few", "more",'not','may' "some", "such", "no", "nor", "too", "very", "can", "will", "just"]
wordcloud1={}
string=""
for i in file_contents:
    if i.isalpha() or i.isspace():
        string+=i
string=string.split()
new=[]
for i in string:
    if i.lower() not in dont_care_terms:
        new.append(i.lower())
for i in new:
    if i.lower() in wordcloud1:
        wordcloud1[i.lower()]+=1
    else:
        wordcloud1[i.lower()]=1

cloud = wordcloud.WordCloud()
cloud.generate_from_frequencies(wordcloud1)
image=cloud.to_array()
```

```
plt.imshow(image, interpolation = 'nearest')
plt.axis('off')
plt.show()
```
